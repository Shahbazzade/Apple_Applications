```python
# Project II
# data was imported and filtered form Kaggle
# https://www.kaggle.com/ramamet4/app-store-apple-data-set-10k-apps/download
import pandas as pd 
import os
import matplotlib.pyplot as plt 
import numpy as np
```

```python
#reading AppleStore.csv file
app = pd.read_csv('AppleStore.csv', index_col=0, encoding = "ISO-8859-1") 
```

```python
#dsiplaying data in app dataset
app
```

```python
#printing top 10 applications by total rating
top_10 = app.sort_values(by = "rating_count_tot", ascending=False)
top_10_app = top_10.head(10)
top_10_app
```

```python
#graphical representation of top 10 applications by total rating 
bar_top_10_app = top_10_app.plot.bar(x="app_name", y="rating_count_tot",title="Top 10 applications")
```

```python
#printing top 5 applications by size
top_5 = app.sort_values(by = "size_bytes", ascending=False)
top_5_app_bysize = top_5.head()
top_5_app_bysize
```

```python
#graphical representation of top 5 applications by size
bar_top_5_app = top_5_app_bysize.plot.bar(x="app_name", y="size_bytes",title="Top 5 applications by size")
```

```python
#printing 5 least size applications 
top_5_low_cap = app.sort_values(by = "size_bytes", ascending=True)
top_5_low_cap_bysize = top_5_low_cap.head()
top_5_low_cap_bysize
```

```python
#graphical representation of top 5 applications with small sizes
bar_top_5_app_small_size = top_5_low_cap_bysize.plot.bar(x="app_name", y="size_bytes",title="Top 5 applications by size")
```

```python
#the 5 most expensive apps
most_expensive_app = app.sort_values(by = "price", ascending=False)
top_5_exp_app = most_expensive_app.head()
top_5_exp_app
```

```python
#graphical representation of top 5 expencive applications
bar_top_5_exp_app = top_5_exp_app.plot.bar(x="app_name", y="price",title="Top 5 expencive applications")
```
